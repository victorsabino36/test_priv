```python
client = bigquery.Client()

dataset = dataset_id

tables_schema = {
    "entregas": [
        bigquery.SchemaField("id_entrega", "INTEGER", mode="REQUIRED"),
        bigquery.SchemaField("data_envio", "DATE"),
        bigquery.SchemaField("data_entrega", "DATE"),
        bigquery.SchemaField("status", "STRING"),
        bigquery.SchemaField("cliente", "STRING"),
        bigquery.SchemaField("cidade", "STRING"),
        bigquery.SchemaField("estado", "STRING"),
        bigquery.SchemaField("valor", "FLOAT"),
        bigquery.SchemaField("tempo_entrega", "STRING"),
    ],
    "veiculos": [
        bigquery.SchemaField("id_veiculo", "INTEGER", mode="REQUIRED"),
        bigquery.SchemaField("placa", "STRING"),
        bigquery.SchemaField("modelo", "STRING"),
        bigquery.SchemaField("capacidade", "FLOAT"),
        bigquery.SchemaField("motorista", "STRING"),
        bigquery.SchemaField("status", "STRING"),
        bigquery.SchemaField("ultima_manutencao", "DATE"),
    ],
    "motoristas":[
        bigquery.SchemaField("id_motorista", "INTEGER", mode="REQUIRED"),
        bigquery.SchemaField("nome", "STRING"),
        bigquery.SchemaField("cnh", "STRING"),
        bigquery.SchemaField("status", "STRING"),
        bigquery.SchemaField("total_entregas", "INTEGER"),
        bigquery.SchemaField("media_atraso", "FLOAT"),
    ],
    "ocorrencias": [
        bigquery.SchemaField("id_motorista", "INTEGER", mode="REQUIRED"),
        bigquery.SchemaField("nome", "STRING"),
        bigquery.SchemaField("cnh", "STRING"),
        bigquery.SchemaField("status", "STRING"),
        bigquery.SchemaField("total_entregas", "INTEGER"),
        bigquery.SchemaField("media_atraso", "FLOAT"),
    ],
}

for table_name, schema in tables_schema.items():
    table_id = f"{dataset_id}.{table_name}"
    table = bigquery.Table(table_id, schema=schema)

    try:
        client.create_table(table)
        print(f"Tabela {table_name} criada com sucesso!")
    except Exception as e:
        print(f"Erro ao criar {table_name}: {e}")
```
