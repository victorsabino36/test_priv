```python
import requests
import pandas as pd
```

```python
url = "https://api.football-data.org/v4/competitions/CL/matches"
headers = {
    'X-Auth-Token': '6d33e6eb2572404abe333f7e013c0903'
}

response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    data = response.json()
    matches = data['matches']
else:
    print(f"Erro na requisição: {response.status_code}")
```

```python
df = pd.DataFrame(matches)
```

```python
df.head() # Exibe as primeiras 5 linhas
```

```python
 #Converter a coluna 'utcDate' para datetime, tratando erros
df['utcDate'] = pd.to_datetime(df['utcDate'], errors='coerce')

# Remover linhas com datas inválidas (NaT)
df = df.dropna(subset=['utcDate'])

# Filtrar os jogos de fevereiro de 2025
jogos_fevereiro_2025 = df[(df['utcDate'].dt.year == 2025) & (df['utcDate'].dt.month == 2)]
jogos_fevereiro_2025_selec = jogos_fevereiro_2025 [['utcDate', 'score']]
```

```python
# Exibir os jogos filtrados
print(jogos_fevereiro_2025_selec['score'])
```
